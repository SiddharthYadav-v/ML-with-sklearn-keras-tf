Single node perceptron

```python
import numpy as np
```

```python
class perceptron:
    def __init__(self, eta=0.1, n_epochs=1000, random_state=None):
        self.eta = eta
        self.n_epochs = n_epochs
        self.random_state = random_state

    def fit(self, X, y):
        self.b = 0
        n = X.shape[1]
        self.w = np.zeros((n))

        for _ in range(self.n_epochs):
            for i in range(len(X)):
                y_pred = self.decision_function(X[i])

                for j in range(n):
                    self.w[j] = self.w[j] + self.eta * (y[i] - y_pred) * X[i][j]
                    self.b = self.b + self.eta * (y[i] - y_pred)

    def decision_function(self, x):
        f = np.dot(self.w, x)
        if f >= self.b:
            return 1
        return 0
    
    def predict(self, X):
        y_pred = []

        for i in range(len(X)):
            y_pred.append(self.decision_function(X[i]))

        return np.array(y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

blob_centers = np.array([[0.2, 2.3], [-2.8, 1.8]])
X, y = make_blobs(n_samples=1000, n_features=2, centers=blob_centers)
```

```python
def plot_clusters(X, y=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1)
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")

plt.figure(figsize=(8, 4))
plot_clusters(X, y)
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                    random_state=42)
```

```python
perc = perceptron(random_state=42)
perc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = perc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
def plot_perceptron_boundary(model):
    # define a mesh grid
    h = 0.01
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), 
                         np.arange(y_min, y_max, h))
    
    # make prediction on the mesh grid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # create a contour plot
    plt.contour(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.8)

    # plot the original data points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')

    # set plot limits and labels 
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")

    # show the plot
    plt.show()
```

```python
plot_perceptron_boundary(perc)
```

Single layer perceptron

```python
class perceptron:
    def __init__(self, eta=0.1, n_epochs=1000, n_outs=10, random_state=None):
        self.eta = eta
        self.n_epochs = n_epochs
        self.n_outs = n_outs
        self.random_state = random_state

    def fit(self, X, y):
        self.b = np.zeros((self.n_outs))
        n = X.shape[1]
        self.w = np.zeros((n, self.n_outs))

        for _ in range(self.n_epochs):
            for k in range(len(X)):
                y_preds = self.decision_function(X[k])

                for i in range(n):
                    for j in range(self.n_outs):
                        self.w[i][j] = self.w[i][j] + self.eta * (y[k] - y_preds[j]) * X[k][i]
                        self.b[j] = self.b[j] + self.eta * (y[k] - y_preds[j])


    def decision_function(self, x):
        f = np.dot(self.w.T, x)
        
        # res = []
        # for i in range(len(f)):
        #     if f[i] > self.b[i]:
        #         res.append(1)
        #     else:
        #         res.append(0)

        # return np.array(res)

        return f + self.b
    
    def predict(self, X):
        y_pred = []

        for i in range(len(X)):
            preds = self.decision_function(X[i])
            # y_pred.append(np.bincount(preds).argmax())
            y_pred.append(preds.argmax())

        return np.array(y_pred)        
```

```python
perc_clf = perceptron(n_outs=2, random_state=42)
perc_clf.fit(X_train, y_train)
```

```python
y_pred = perc_clf.predict(X_train)
accuracy_score(y_test, y_pred)
```
